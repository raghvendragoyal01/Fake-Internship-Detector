```python
import pandas as pd
import re

df = pd.read_csv("cleaned_Final_dataset.csv")
```

```python
df["char_count"] = df["combined_text"].astype(str).apply(len)
df["word_count"] = df["combined_text"].astype(str).apply(
    lambda x: len(x.split()))
df["sentence_count"] = df["combined_text"].astype(str).apply(
    lambda x: len(re.findall(r"[.!?]", x)) + 1)
df["avg_word_length"] = df["combined_text"].astype(str).apply(
    lambda x: sum(len(w) for w in x.split()) / max(len(x.split()), 1))
```

```python
def uppercase_ratio(text):
    text = str(text)
    total = len(text)
    upper = sum(1 for c in text if c.isupper())
    return upper / max(total, 1)
df["uppercase_ratio"] = df["combined_text"].apply(
    uppercase_ratio)
```

```python
def digit_ratio(text):
    text = str(text)
    digits = sum(1 for c in text if c.isdigit())
    return digits / max(len(text), 1)
df["digit_ratio"] = df["combined_text"].apply(
    digit_ratio)
```

```python
def special_ratio(text):
    text = str(text)
    special = len(
        re.findall(r"[^a-zA-Z0-9\s]", text)
    )
    return special / max(len(text), 1)
df["special_ratio"] = df["combined_text"].apply(
    special_ratio
)
```

```python
df["exclamation_count"] = (
    df["combined_text"]
    .astype(str)
    .str.count("!")
)
```

```python
df["question_count"] = (
    df["combined_text"]
    .astype(str)
    .str.count(r"\?")
)
```

```python
df.to_csv(
    "text_feature_dataset.csv",
    index=False
)
```
