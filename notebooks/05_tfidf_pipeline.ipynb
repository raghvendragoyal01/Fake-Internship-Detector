```python
import pandas as pd
import joblib

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv("feature_dataset_v3.csv")
```

```python
df["combined_text"] = (
    df["combined_text"]
    .fillna("")
    .astype(str)
)
```

```python
tfidf = TfidfVectorizer(
    max_features=5000,
    stop_words="english",
    ngram_range=(1,2),
    min_df=2,
    max_df=0.95,
    sublinear_tf=True)

X_tfidf = tfidf.fit_transform(
    df["combined_text"]
)
```

```python
joblib.dump(
    X_tfidf,
    "tfidf_matrix.pkl"
)
```

```python
joblib.dump(
    tfidf,
    "tfidf_vectorizer.pkl"
)
```

```python
print()
print("Vocabulary Size :")
print(len(tfidf.vocabulary_))
print()
print("TFIDF Shape :")
print(X_tfidf.shape)
```

```python
import pandas as pd

# Load Dataset
df = pd.read_csv("feature_dataset_v3.csv")

drop_cols = [
    "has_bitly",
    "has_blogspot",
    "has_wordpress",
    "has_weebly"
]

df.drop(
    columns=drop_cols,
    inplace=True,
    errors="ignore"
)


df["avg_sentence_length"] = df[
    "avg_sentence_length"
].clip(upper=100)

df.to_csv(
    "feature_dataset_final.csv",
    index=False
)

print(df.shape)
print(df.columns.tolist())
```
