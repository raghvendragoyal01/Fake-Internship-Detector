# 📊 Dataset Inspection & Exploratory Data Analysis (EDA)

## 🎯 Objective

The objective of this notebook is to perform an initial inspection and exploratory analysis of the collected internship and job listing dataset. This step helps in understanding the structure, quality, and distribution of the data before preprocessing and model development.

---

## 📌 Tasks Performed

* Loaded the Indian Job Listings dataset.
* Examined dataset dimensions and feature information.
* Inspected data types and column names.
* Checked for missing values across all features.
* Identified duplicate records.
* Analyzed the target variable distribution (Real vs Fake Jobs).
* Examined salary information and text-based features.
* Generated initial insights for further preprocessing.

---

## 🔍 Key Findings

* The dataset contains **6,553 job listings** with **13 features**.
* A significant class imbalance exists between **Real Jobs** and **Fake Jobs**.
* The `Salary_Amount` column contains approximately **75% missing values**.
* Most textual features such as Job Title and Job Description are complete and suitable for NLP-based feature extraction.
* Dataset inspection revealed the need for additional fake job samples to improve model performance.

---

## 🚀 Outcome

The notebook successfully identifies data quality issues and provides the foundation for dataset augmentation, preprocessing, and machine learning model development in subsequent notebooks.

3. Missing Values Checking 


```python
missing = pd.DataFrame({
    "Missing Count": df.isnull().sum(),
    "Missing %": (df.isnull().sum()/len(df))*100
})

missing.sort_values(
    by="Missing %",
    ascending=False
)
```
