```python
import pandas as pd
import joblib

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import csr_matrix, hstack
```

```python
df = pd.read_csv("feature_dataset_final.csv")
```

```python
HANDCRAFTED = [
    "char_count",
    "word_count",
    "sentence_count",
    "avg_word_length",
    "digit_ratio",
    "special_ratio",
    "exclamation_count",
    "question_count",
    "has_email",
    "gmail_email",
    "yahoo_email",
    "outlook_email",
    "salary_present",
    "salary_daily",
    "salary_monthly",
    "salary_hourly",
    "avg_sentence_length",
    "unique_word_ratio",
    "repetition_score",
    "long_word_ratio",
    "numeric_token_ratio"
]
```

```python
train_df, test_df = train_test_split(
    df,
    test_size=0.2,
    random_state=42,
    stratify=df["label"]
)

print()
print(train_df.shape)
print(test_df.shape)
```

```python
tfidf = TfidfVectorizer(
    max_features=5000,
    stop_words="english",
    ngram_range=(1,2),
    min_df=2,
    max_df=0.95,
    sublinear_tf=True
)

X_train_text = tfidf.fit_transform(
    train_df["combined_text"]
)

X_test_text = tfidf.transform(
    test_df["combined_text"]
)
```

```python
X_train_manual = csr_matrix(
    train_df[HANDCRAFTED].values
)

X_test_manual = csr_matrix(
    test_df[HANDCRAFTED].values
)
```

```python
X_train = hstack([
    X_train_text,
    X_train_manual
])

X_test = hstack([
    X_test_text,
    X_test_manual
])
```

```python
y_train = train_df["label"]
y_test = test_df["label"]
```

```python
joblib.dump(
    tfidf,
    "tfidf_vectorizer.pkl"
)

joblib.dump(
    X_train,
    "X_train.pkl"
)

joblib.dump(
    X_test,
    "X_test.pkl"
)

joblib.dump(
    y_train,
    "y_train.pkl"
)

joblib.dump(
    y_test,
    "y_test.pkl"
)

print()
print("Train Matrix :", X_train.shape)
print("Test Matrix :", X_test.shape)
print()
print(y_train.value_counts())
print()
print(y_test.value_counts())
```
