```python
import pandas as pd
import re

df = pd.read_csv("text_feature_dataset.csv")

```

```python
def contains_email(text):

    return int(bool(
        re.search(
            r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}",
            str(text)
        )
    ))
df["has_email"] = df["combined_text"].apply(
    contains_email
)
```

```python
df["gmail_email"] = (
    df["combined_text"]
    .str.contains("@gmail.com", case=False)
    .fillna(False)
    .astype(int)
)
```

```python
df["yahoo_email"] = (
    df["combined_text"]
    .str.contains("@yahoo", case=False)
    .fillna(False)
    .astype(int)
)
```

```python
df["outlook_email"] = (
    df["combined_text"]
    .str.contains("@outlook", case=False)
    .fillna(False)
    .astype(int)
)
```

```python
df["has_bitly"] = (
    df["combined_text"]
    .str.contains("bit.ly", case=False)
    .fillna(False)
    .astype(int)
)

df["has_blogspot"] = (
    df["combined_text"]
    .str.contains("blogspot", case=False)
    .fillna(False)
    .astype(int)
)

df["has_wordpress"] = (
    df["combined_text"]
    .str.contains("wordpress", case=False)
    .fillna(False)
    .astype(int)
)

df["has_weebly"] = (
    df["combined_text"]
    .str.contains("weebly", case=False)
    .fillna(False)
    .astype(int)
)
```

```python
df["salary_present"] = (
    df["salary"]
    .astype(str)
    .str.contains(r"\d")
    .astype(int)
)

df["salary_daily"] = (
    df["salary"]
    .astype(str)
    .str.contains("day", case=False)
    .astype(int)
)

df["salary_monthly"] = (
    df["salary"]
    .astype(str)
    .str.contains("month", case=False)
    .astype(int)
)

df["salary_hourly"] = (
    df["salary"]
    .astype(str)
    .str.contains("hour", case=False)
    .astype(int)
)

```

```python
df.to_csv(
    "feature_dataset_v2.csv",
    index=False
)
```
